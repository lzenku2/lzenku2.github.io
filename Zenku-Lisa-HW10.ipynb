### Creating the Plots

```python
brush = alt.selection_interval(encodings=['x','y'])
```

```python
chart1 = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/building_inventory.csv"},
  "mark":"bar",
  "encoding":{
    "x":{"field":"Agency Name", "type":"nominal"},
    "y":{"field":"Congressional Full Name","type":"nominal"}, 
    "color":{"aggregate":"count","type":"quantitative"}
  }
}).add_selection(
    brush
)

chart1
```

```python
chart2 = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/building_inventory.csv"},
  "mark":"bar",
  "encoding":{
    "x":{"bin":True, "field":"Square Footage", "type":"quantitative", "title":"Square Footage"},
    "y":{
      "aggregate":"count", 
      "type":"quantitative",
      "axis":{"title":"Count"}
    } 
  } 
}).transform_filter(
    brush
)

chart2
```

### FINAL PLOT

```python
chart = chart1 | chart2
chart
```

```python
chart.properties().save("hw10.json")
```
